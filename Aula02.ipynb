```python
numero_secreto = 7
tentativas = 5

while tentativas > 0:
    palpite = int(input("Adivinhe o número (entre 1 e 10): "))

    if palpite == numero_secreto:
        print("Parabéns! Você acertou o número secreto.")
        break
    elif palpite > numero_secreto:
        print("Tente um número menor.")
    else:
        print("Tente um número maior.")

    tentativas -= 1
    print(f"Você ainda tem {tentativas} tentativa(s) restante(s).")

if tentativas == 0:
    print("Você perdeu! O número secreto era:", numero_secreto)
```

```python
relatorio = "Relatório de Compras:\n"
relatorio += f"{'Produto':<15}{'Quantidade':<12}{'Preço Unitário':<15}{'Total':<10}\n"
relatorio += "-" * 50 + "\n"

while True:
    produto = input("Digite o nome do produto (ou 'sair' para encerrar): ").strip()
    if produto.lower() == 'sair':
        break

    quantidade = int(input(f"Digite a quantidade de {produto}: "))
    preco_unitario = float(input(f"Digite o preço unitário de {produto} (em R$): "))

    total = quantidade * preco_unitario
    relatorio += f"{produto:<15}{quantidade:<12}{preco_unitario:<15.2f}{total:<10.2f}\n"

print("\n" + relatorio)
```
